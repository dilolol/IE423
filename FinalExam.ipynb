### Explained Variance by Number of Principal Components

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.impute import SimpleImputer # Import the imputer
import matplotlib.pyplot as plt
import numpy as np

# Select only numerical columns for PCA
numerical_columns = df.select_dtypes(include=[np.number]).columns
X = df[numerical_columns]

# Impute missing values with the mean of each column
imputer = SimpleImputer(strategy='mean') # Create an imputer instance
X_imputed = imputer.fit_transform(X) # Impute the missing values

# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_imputed) # Use the imputed data

# Apply PCA
pca = PCA()
X_pca = pca.fit_transform(X_scaled)

# Explained variance
explained_variance = pca.explained_variance_ratio_

# Plot the explained variance
plt.figure(figsize=(10, 7))
plt.plot(range(1, len(explained_variance) + 1), np.cumsum(explained_variance), marker='o', linestyle='--')
plt.title('Explained Variance by Number of Principal Components')
plt.xlabel('Number of Principal Components')
plt.ylabel('Cumulative Explained Variance')
plt.grid(True)  # Ensure grid is visible
plt.show()

# Print the explained variance
print("Explained Variance for each component:")
for i, variance in enumerate(explained_variance, start=1):
    print(f"Component {i}: {variance:.4f}")
```


- **Top Contributors**: The first three components capture 47.39% of the variance:
  - **Component 1**: 24.35% — Represents a significant portion of data variance.
  - **Component 2**: 12.04%
  - **Component 3**: 10.00%
- **Significant Accumulation**: Components 4 to 10 progressively add less variance each, but cumulatively contribute to detailing the data's underlying structure.

### Outliers

```python
# Create df_encoded after encoding categorical features
df_encoded = df.copy()  # Create a copy to avoid modifying the original DataFrame

# Identify outliers using the IQR method for a specific column
# Replace 'Revenue' with the actual column name if needed
column_name = 'Revenue'

# Check the data type of the 'Revenue' column
print(df_encoded[column_name].dtype)

# If it's boolean, convert it to a suitable numerical type (e.g., int)
if df_encoded[column_name].dtype == 'bool':
    df_encoded[column_name] = df_encoded[column_name].astype(int)

Q1 = df_encoded[column_name].quantile(0.25)
Q3 = df_encoded[column_name].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = df_encoded[(df_encoded[column_name] < lower_bound) | (df_encoded[column_name] > upper_bound)]

print(outliers)
```

### Cluster Analysis

#### For Administrative_Duration, Informational_Duration and ProductRelated_Duration


```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
X_cluster = df[['Administrative_Duration', 'Informational_Duration', 'ProductRelated_Duration']]

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_cluster)

# Using the Elbow method to find the optimal number of clusters
inertia = []
for k in range(1, 10):
    kmeans = KMeans(n_clusters=k, random_state=1)
    kmeans.fit(X_scaled)
    inertia.append(kmeans.inertia_)

plt.figure(figsize=(8, 4))
plt.plot(range(1, 10), inertia, marker='o')
plt.title('Elbow Method For Optimal k')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.show()

# Let's assume the optimal number of clusters is 3
kmeans = KMeans(n_clusters=3, random_state=1)
kmeans.fit(X_scaled)
df['cluster'] = kmeans.labels_
```

```python
import matplotlib.pyplot as plt
# Model names
models = ['Logistic Regression', 'Decision Tree', 'Random Forest', 'SVM', 'XGBoost', 'KNN']

# Accuracy before Hyperparameter Tuning (replace with your actual baseline accuracies)
accuracy_before = [0.89, 0.85, 0.90, 0.88, 0.89, 0.87]

# Accuracy after Hyperparameter Tuning (from your results)
accuracy_after = [0.8962, 0.8948, 0.9111, 0.8970, 0.8994, 0.8889]

# Calculate the improvement
improvement = [after - before for before, after in zip(accuracy_before, accuracy_after)]

# Set up the figure and axes
fig, ax = plt.subplots()

# Bar plot for improvement
ax.bar(models, improvement, color='skyblue')

# Add labels and title
ax.set_xlabel('Models')
ax.set_ylabel('Improvement in Accuracy')
ax.set_title('Improvement in Accuracy after Hyperparameter Tuning')
ax.set_xticklabels(models, rotation=45, ha='right')

# Annotate bars with the improvement values
for i, v in enumerate(improvement):
    ax.text(i, v + 0.005, str(round(v, 4)), ha='center', va='bottom')

# Show the plot
plt.tight_layout()
plt.show()
```

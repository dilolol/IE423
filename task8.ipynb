```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

user_ratings = pd.merge(user_ratings_1, user_ratings_2, on='JokeId')
user_ratings.set_index('JokeId', inplace=True)

user_ratings_filled = user_ratings.apply(lambda row: row.fillna(row.mean()), axis=1)

joke_text['JokeText'] = joke_text['JokeText'].fillna('')

tfidf_vectorizer = TfidfVectorizer(analyzer='word', ngram_range=(1, 2), min_df=1, stop_words='english')
tfidf_matrix = tfidf_vectorizer.fit_transform(joke_text['JokeText'])

cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)


joke_ids = joke_text['JokeId']
indices = pd.Series(joke_text.index, index=joke_text['JokeId'])
```

```python
def get_recommendations(joke_id, num_recommendations=5):
    idx = indices[joke_id]
    sim_scores = list(enumerate(cosine_sim[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:num_recommendations+1]
    joke_indices = [i[0] for i in sim_scores]
    return joke_text.iloc[joke_indices]['JokeText']

recommended_jokes = get_recommendations(0)
print(recommended_jokes)
```

# Collaborative Filtering


```python
def get_collaborative_recommendations(user_id, num_recommendations=5):
    # Access row by position using iloc since the index is not sequential
    user_ratings = user_ratings_filled.iloc[user_id]
    unrated_jokes = user_ratings[user_ratings.isnull()]

    # Handle the case where all jokes are rated
    if unrated_jokes.empty:
        return "User has rated all jokes"
    else:
      print(unrated_jokes)

    item_similarity = cosine_similarity(user_ratings_filled.T)

    joke_scores = np.dot(item_similarity, unrated_jokes.values)
    top_indices = np.argsort(joke_scores)[::-1][:num_recommendations]
    recommended_jokes = user_ratings_filled.columns[top_indices]
    return recommended_jokes

# Example usage
user_id = 1
recommended_jokes = get_collaborative_recommendations(user_id)
print(recommended_jokes)
```

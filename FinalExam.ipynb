### Explained Variance by Number of Principal Components

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import numpy as np

# Select only numerical columns for PCA
numerical_columns = df.select_dtypes(include=[np.number]).columns
X = df[numerical_columns]

# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Apply PCA
pca = PCA()
X_pca = pca.fit_transform(X_scaled)

# Explained variance
explained_variance = pca.explained_variance_ratio_

# Plot the explained variance
plt.figure(figsize=(10, 7))
plt.plot(range(1, len(explained_variance) + 1), np.cumsum(explained_variance), marker='o', linestyle='--')
plt.title('Explained Variance by Number of Principal Components')
plt.xlabel('Number of Principal Components')
plt.ylabel('Cumulative Explained Variance')
plt.grid(True)  # Ensure grid is visible
plt.show()

# Print the explained variance
print("Explained Variance for each component:")
for i, variance in enumerate(explained_variance, start=1):
    print(f"Component {i}: {variance:.4f}")
```


- **Top Contributors**: The first three components capture 47.39% of the variance:
  - **Component 1**: 24.35%  Represents a significant portion of data variance.
  - **Component 2**: 12.04%
  - **Component 3**: 10.00%
- **Significant Accumulation**: Components 4 to 10 progressively add less variance each, but cumulatively contribute to detailing the data's underlying structure.
- **Diminishing Returns**: Components 11 to 14 capture minimal additional variance, indicating less relevance.

### Encode categorical variables and normalize/standardize numerical features:

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, StandardScaler

# Identify categorical columns
categorical_cols = ['Month', 'OperatingSystems', 'Browser', 'Region', 'TrafficType', 'VisitorType', 'Weekend']

# Check for unique values in the Month column
print("Unique values in 'Month' before mapping:", df['Month'].unique())

# Fill missing values in Month if necessary
df['Month'] = df['Month'].fillna('Unknown')

# Define the mapping
month_mapping = {'Jan': 1, 'Feb': 2, 'Mar': 3, 'Apr': 4, 'May': 5, 'Jun': 6, 'Jul': 7, 'Aug': 8, 'Sep': 9, 'Oct': 10, 'Nov': 11, 'Dec': 12}

# Map the Month values
df['Month'] = df['Month'].map(month_mapping)

# Check for any NaN values in the Month column
print("NaN values in 'Month' after mapping:", df['Month'].isna().sum())

# Encode other categorical variables using Label Encoding
label_encoders = {}
for col in categorical_cols:
    if col != 'Month':  # Month is already encoded
        le = LabelEncoder()
        df[col] = le.fit_transform(df[col])
        label_encoders[col] = le  # Store encoder to inverse transform if needed

# Define features and target variable
X = df.drop('Revenue', axis=1)
y = df['Revenue']

# Standardize numerical features
scaler = StandardScaler()
numerical_columns = X.select_dtypes(include=[np.number]).columns
X[numerical_columns] = scaler.fit_transform(X[numerical_columns])

# Display the month to check
print(df[['Month']].head())
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_auc_score
from sklearn.preprocessing import StandardScaler

# Define features and target variable
X = df.drop('Revenue', axis=1)  # Ensure all new features are included
y = df['Revenue']

# Split the dataset into training (80%) and testing (20%) sets
X_train, X_test, y_train, y_test = train_test_split(X, y,train_size=0.8, test_size=0.2,random_state=1)
```

### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
from sklearn.impute import SimpleImputer

# Define features and target variable
X = df.drop('Revenue', axis=1)
y = df['Revenue']

# Split the dataset into training (80%) and testing (20%) sets
X_train, X_test, y_train, y_test = train_test_split(X, y,train_size=0.8, test_size=0.2,random_state=1)

# Handle missing values using imputation (replace NaNs with the mean of the column)
imputer = SimpleImputer(strategy='mean')
X_train = imputer.fit_transform(X_train)
X_test = imputer.transform(X_test)  # Use the same imputer fitted on training data

# Initialize and train the Logistic Regression model
logreg = LogisticRegression(max_iter=1000)
logreg.fit(X_train, y_train)

# Predict on the testing set
y_pred_logreg = logreg.predict(X_test)

# Evaluate the model
print("Logistic Regression Performance:")
print(classification_report(y_test, y_pred_logreg))
```

```python
y_proba_logreg = logreg.predict_proba(X_test)[:, 1]  # Probability of the positive class

# Evaluate the model using ROC-AUC
roc_auc_logreg = roc_auc_score(y_test, y_proba_logreg)
print(f"ROC-AUC for Logistic Regression: {roc_auc_logreg:.4f}")
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

# Compute confusion matrix for Logistic Regression
cm_lr = confusion_matrix(y_test, y_pred_logreg)

# Create a ConfusionMatrixDisplay object
disp_lr = ConfusionMatrixDisplay(confusion_matrix=cm_lr, display_labels=['No Purchase', 'Purchase'])

# Plot the confusion matrix
plt.figure(figsize=(10, 7))
disp_lr.plot(cmap='Blues', values_format='d', ax=plt.gca(), colorbar=False)
plt.title('Confusion Matrix for Logistic Regression')
plt.show()
```

### Decision Trees:

```python
from sklearn.tree import DecisionTreeClassifier

# Initialize and train the Decision Tree model
dtree = DecisionTreeClassifier(random_state=42)
dtree.fit(X_train, y_train)

# Predict on the testing set
y_pred_dtree = dtree.predict(X_test)

# Evaluate the model
print("Decision Tree Performance:")
print(classification_report(y_test, y_pred_dtree))
```

```python
y_proba_dtree = dtree.predict_proba(X_test)[:, 1] # Calculate predicted probabilities
roc_auc_dtree = roc_auc_score(y_test, y_proba_dtree)
print(f"ROC-AUC for Decision Tree: {roc_auc_dtree:.4f}")
```

```python
# Compute confusion matrix for Decision Tree
cm_dt = confusion_matrix(y_test, y_pred_dtree)

# Create a ConfusionMatrixDisplay object
disp_dt = ConfusionMatrixDisplay(confusion_matrix=cm_dt, display_labels=['No Purchase', 'Purchase'])

# Plot the confusion matrix
plt.figure(figsize=(10, 7))
disp_dt.plot(cmap='Blues', values_format='d', ax=plt.gca(), colorbar=False)
plt.title('Confusion Matrix for Decision Tree')
plt.show()
```

### **ROC Curve:**

```python
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, auc
# Define a dictionary to hold models and their predictions
models_predictions = {
    'Logistic Regression': y_proba_logreg,
    'Decision Tree': y_proba_dtree,
    'Random Forest': y_proba_rf,  # Corrected indentation
    'SVM': y_proba_svm,
    'XGBoost': y_proba_xgb,
    'KNN': y_proba_knn
}  # Close the dictionary here

# Initialize the figure
plt.figure(figsize=(10, 7))

# Plot ROC curves for each model
for model_name, y_proba in models_predictions.items():
    fpr, tpr, _ = roc_curve(y_test, y_proba)
    roc_auc = auc(fpr, tpr)
    plt.plot(fpr, tpr, label=f'{model_name} (AUC = {roc_auc:.2f})')
    plt.plot([0, 1], [0, 1], 'k--')  # Diagonal line for random classifier
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('Receiver Operating Characteristic (ROC) Curve')
    plt.legend(loc="lower right")

# Show the plot
plt.show()
```

### Using Hyperparameter Tuning

#### Logistic Regression

- **C:** Regularization strength

- **Solver:** Algorithm to use for optimization

```python
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler # Import the scaler

# Define parameter grid
param_grid_lr = {
    'C': [0.01, 0.1, 1, 10],
    'solver': ['liblinear', 'lbfgs']
}

# Initialize Logistic Regression
logreg = LogisticRegression(max_iter=1000)

# Assuming X_train exists, scale the features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train) # Create the missing variable

# Set up GridSearchCV for hyperparameter tuning
grid_search_lr = GridSearchCV(logreg, param_grid_lr, cv=5, scoring='roc_auc')
grid_search_lr.fit(X_train_scaled, y_train)

print("Best parameters for Logistic Regression:", grid_search_lr.best_params_)
```

```python
# Use the best parameters from GridSearchCV
logreg_tuned = LogisticRegression(C=grid_search_lr.best_params_['C'], solver=grid_search_lr.best_params_['solver'], max_iter=1000)

# Fit the model on the training data
logreg_tuned.fit(X_train_scaled, y_train)

# Predict on the test data
y_pred_lr_tuned = logreg_tuned.predict(scaler.transform(X_test))  # Scale the test data using the same scaler

# Calculate accuracy
accuracy_lr_tuned = accuracy_score(y_test, y_pred_lr_tuned)
print(f"Accuracy for Logistic Regression after tuning: {accuracy_lr_tuned:.4f}")
```

```python
from sklearn.tree import DecisionTreeClassifier

# Define parameter grid
param_grid_dt = {
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10]
}

# Initialize Decision Tree
dtree = DecisionTreeClassifier(random_state=42)

# Set up GridSearchCV
grid_search_dt = GridSearchCV(dtree, param_grid_dt, cv=5, scoring='roc_auc')
grid_search_dt.fit(X_train, y_train)

print("Best parameters for Decision Tree:", grid_search_dt.best_params_)
```
